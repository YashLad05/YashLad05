```python
df_year1 = pd.read_csv("year1_data.csv")
df_year1
```

```python
print(df_year1.shape)
print(df_year1.size)
print(df_year1.ndim)
df_year1.columns
```

```python
x = (df_year1['tender_type_desc'])
y = (df_year1['store_no'])

plt.figure(figsize = (15,5))
plt.plot(x,y)
plt.xlabel('Tender Description')
plt.ylabel('Store Number')
plt.title('Tender Description on store level')
plt.show()
```

```python
# Purchase Frequency from store level
plt.figure(figsize = (10,5))
plt.hist(df_year1['store_no'])
plt.xlabel('Store Number')
plt.ylabel('Frequency of visit')
plt.title('Purchase Frequency from store level')
plt.show()
```

```python
# Item name on store level
x = (df_year1['item1_desc'])
y = (df_year1['store_no'])

plt.figure(figsize = (15,10))
plt.xlabel('Store nimber')
plt.ylabel('Item Name')
plt.title('Data Plot 1')
plt.plot(y,x)
```

```python
# Year 1 Data Promotions
x = (df_year1['cal_week_nbr'])
y = (df_year1['promotion'])
plt.figure(figsize = (5,3))
plt.plot(x,y)
plt.title('Year 1 Data Promotions')
plt.show()
```

```python
# Sale Frequency on Ethnic Group level
x = (df_year1['ethnic_group'])
y = (df_year1['sales'])
plt.figure(figsize = (10,5))
plt.barh(x,y, color='g')
plt.ylabel('Ethnic Group')
plt.xlabel('Sales')
plt.title('Sales on Ethnic Group level')
plt.show()
```
